```python
from selenium import webdriver
import itertools
import pandas as pd

print('Enter 1 if you want to search for Book,Enter 2 if you want to search for Electronics:: ', end='')
choice = int(input())

print('What do you want to search for ? : ', end='')
query_string = input()
url = "https://www.flipkart.com/search?q=" + query_string + '&page='

driver = webdriver.Chrome("chromedriver")
names = []
prices = []
links = []
i=0
j=0
k=0

for pa in range(1,3):
    furl = ""
    furl = url + str(pa)
    driver.get(furl)
    
   
    if choice == 1:
        for_names= driver.find_elements_by_xpath("//a[@class='_2cLu-l']")
        for_prices= driver.find_elements_by_xpath("//div[@class='_1vC4OE']")
        for_links= driver.find_elements_by_css_selector('._2cLu-l')
        for n in for_names:
            names.append(n.text)
            i += 1

        for p in for_prices:
            prices.append(p.text)
            j +=1

        for l in for_links:
            links.append(l.get_attribute('href'))
            k +=1
            
        

    elif choice == 2:
        for_names= driver.find_elements_by_xpath("//div[@class='_3wU53n']")
        for_prices= driver.find_elements_by_xpath("//div[@class='_1vC4OE _2rQ-NK']")
        for_links= driver.find_elements_by_css_selector('._31qSD5')
        for n in for_names:
            names.append(n.text)
            i += 1

        for p in for_prices:
            prices.append(p.text)
            j +=1

        for l in for_links:
            links.append(l.get_attribute('href'))
            k +=1



df1 = pd.DataFrame({'Name of Product':names,'Price of Product':prices,'links of products':links})
df1
```

```python
df1.to_csv('projectfile.csv',index=False,encoding='utf-8')
```
